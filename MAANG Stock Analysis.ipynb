## Moving Average-MA

A moving average (MA) is a widely used indicator in technical analysis that helps smooth out price action by filtering out the noise from random price fluctuations. It is a trend-following, or lagging, indicator because it is based on past prices.

```python
ma_day = [10,20,30]

for ma in ma_day:
    col_name = "MA for %s days" %(str(ma))
    META[col_name]= pd.DataFrame.rolling(META['close'],ma).mean()
```

Now want to analyze the returns of all the stocks in our list. Let's go ahead and build a DataFrame with all the ['close'] columns for each of the stocks dataframes.

```python
closing_df1 = pd.DataFrame(META['close'])
close1 = closing_df1.rename(columns= {"close":"META_close"})

closing_df2 = pd.DataFrame(AAPL['close'])
close2 = closing_df2.rename(columns= {"close":"AAPL_close"})

closing_df3 = pd.DataFrame(AMZN['close'])
close3 = closing_df3.rename(columns= {"close":"AMZN_close"})

closing_df4 = pd.DataFrame(NFLX['close'])
close4 = closing_df4.rename(columns= {"close":"NFLX_close"})

closing_df5 = pd.DataFrame(GOOG['close'])
close5 = closing_df5.rename(columns= {"close":"GOOG_close"})

closing_df = pd.concat([close1,close2,close3,close4,close5],axis = 1)
closing_df.head()
```

## Now that we have all the closing prices, let's go ahead and get the daily return for all the stocks.

```python
# Make a new daily returns DataFrame

daily_returns = closing_df.pct_change()
```

Now we can compare the daily percentage return of two stocks to check how they are correlated.

```python
# Comparing daily returns of META and AAPL
sns.jointplot('META_close','AAPL_close',daily_returns,kind='scatter',color='seagreen')
```

Python's Seaborn and pandas packages make it very easy to repeat this comparison analysis for every possible combination of stocks in our MAANG stock ticker list. We can use sns.pairplot() to automatically create this plot.

```python
# We can simply call pairplot on our DataFrame for an automatic visual analysis of all the comparisons

sns.pairplot(daily_returns.dropna())
```

From above plot , we can see the daily returns between all stocks

```python
import warnings
warnings.filterwarnings('ignore')
returns_fig = sns.PairGrid(daily_returns.dropna())

returns_fig.map_upper(plt.scatter,color='purple')

returns_fig.map_lower(sns.kdeplot,cmap='cool_d')

returns_fig.map_diag(plt.hist,bins =30)
```

Finally, we could also do a correlation plot, to get actual numerical values for the correlation between the stocks' daily return values.

```python
corr= daily_returns.dropna().corr()

mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)]=True

fig,ax=plt.subplots(figsize=(11,9))

cmap= sns.diverging_palette(220,10,as_cmap=True)

sns.heatmap(corr,mask=mask,cmap=cmap,annot = True)
```
